```python
import pandas as pd

from sklearn.decomposition import PCA

from sklearn.preprocessing import StandardScaler

import matplotlib.pyplot as plt

df = pd.read_csv("Iris.csv")
```

```python
data = pd.read_csv("Iris.csv")
```

```python
numerical_features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']  # Assuming these are your numerical columns
numerical_data = data[numerical_features]
```

```python
scaler = StandardScaler()
```

```python
scaled_data = scaler.fit_transform(numerical_data)
```

```python
pca = PCA(n_components=2)
```

```python
reduced_data = pca.fit_transform(scaled_data)
```

```python
species_mapping = {'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2}
color_indices = data['Species'].map(species_mapping)
```

```python
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=color_indices)
plt.xlabel("PC 1")
plt.ylabel("PC 2")
plt.title("PCA Visualization")
plt.show()
```
